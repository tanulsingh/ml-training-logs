**Why is this better than GQA?** _MLA stores `d_c` floats per token (e.g., 512). GQA stores `2 * n_kv_heads * d_head` floats per token (e.g., 2 * 8 * 128 = 2048). Yet MLA gives every head its own reconstructed K, V — full per-head diversity. The compression is per-token, not per-head._

**The two things missing here.** _Production MLA includes (a) the **absorbed up-projection trick** — fold W_k_up into W_q so attention scores become `(q_absorbed) · c^T` instead of `q · (W_k_up · c)^T`, eliminating the need to ever materialize the full K vector. And (b) **decoupled RoPE**: position info is added on a separate small key (since RoPE breaks the absorption trick when applied to the compressed latent). These are the engineering details that turn the principle into a production-grade kernel._
