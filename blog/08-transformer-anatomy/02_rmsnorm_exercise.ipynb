# RMSNorm — Exercise

Companion to [Anatomy § Normalization](https://tanulsingh.github.io/blog/transformer-anatomy#normalization-the-training-stabilizer).

**LayerNorm** (2016): subtract mean, divide by std, learned scale + bias.

$$\text{LayerNorm}(x) = \gamma \odot \frac{x - \mu}{\sqrt{\sigma^2 + \epsilon}} + \beta$$

**RMSNorm** (2019): drop the mean subtraction and the bias. Just divide by RMS.

$$\text{RMSNorm}(x) = \gamma \odot \frac{x}{\sqrt{\text{mean}(x^2) + \epsilon}}$$

Why drop them? Mean subtraction is a centering shift that the next layer's weights can absorb anyway — it's not what stabilizes training. The magnitude division is. RMSNorm keeps the essential part and runs ~10-15% faster.

Used by every modern LLM: Llama, Mistral, Gemma, Qwen, DeepSeek.
